## Data cleaning and Visualizing

```python
# if there is any missing data in the dataset remove it
df.dropna(inplace=True)
df.info()
```

```python
# get the target variable
df['Primary Type'].value_counts()
# print the target variable as plot 
plt.figure(figsize=(10,10))
sns.countplot(y='Primary Type', data=df, order = df['Primary Type'].value_counts().index)
plt.show()
```

```python
# make the less than 200 count as other
threshold = 200
crime_types = df['Primary Type'].value_counts()
df['Primary Type'] = np.where(df['Primary Type'].isin(crime_types.index[crime_types < threshold]), 'OTHER', df['Primary Type'])
df['Primary Type'].value_counts()

# print the target variable as plot
plt.figure(figsize=(10,10))
sns.countplot(y='Primary Type', data=df, order = df['Primary Type'].value_counts().index)
plt.show()
```

```python
# if there is any missing data in the dataset remove it
df.dropna(inplace=True)
# Remove irrelevant/not meaningfull attributes
df = df.drop(['Unnamed: 0'], axis=1)
df = df.drop(['ID'], axis=1)
df = df.drop(['Case Number'], axis=1) 

df.info()
```

```python
# Splitting the Date to Day, Month, Year, Hour, Minute, Second
df['date2'] = pd.to_datetime(df['Date'])
df['Year'] = df['date2'].dt.year
df['Month'] = df['date2'].dt.month
df['Day'] = df['date2'].dt.day
df['Hour'] = df['date2'].dt.hour
df['Minute'] = df['date2'].dt.minute
df['Second'] = df['date2'].dt.second 
df = df.drop(['Date'], axis=1) 
df = df.drop(['date2'], axis=1) 
df = df.drop(['Updated On'], axis=1)
df.head()
```

```python
# number of unique values in target variable
print("Number of Crime Types =", df['Primary Type'].nunique()) # Number of Classes = 8
```

```python
# Classes
classes = df['Primary Type'].unique()
print(classes)
```

**Encoding and factorizing**

```python
# Encoding the target variable from 0 to 7
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Primary Type'] = le.fit_transform(df['Primary Type'])
df['Primary Type'].value_counts()
```

```python
df.head()
```

```python
# Convert Categorical Attributes to Numerical
df['Block'] = pd.factorize(df["Block"])[0]
df['IUCR'] = pd.factorize(df["IUCR"])[0]
df['Description'] = pd.factorize(df["Description"])[0]
df['Location Description'] = pd.factorize(df["Location Description"])[0]
df['FBI Code'] = pd.factorize(df["FBI Code"])[0]
df['Location'] = pd.factorize(df["Location"])[0]

df.head()
```

**Remove Outliers**

```python
def plot_histograms(df):
    numerical_columns = df.select_dtypes(exclude=['bool']).columns
    for column in numerical_columns:
        plt.figure(figsize=(8, 6))
        sns.histplot(df[column], kde=True)
        plt.title(f'Histogram of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()

# Function to plot boxplots for numerical columns
def plot_boxplots(df):
    numerical_columns = df.select_dtypes(exclude=['bool']).columns
    for column in numerical_columns:
        plt.figure(figsize=(8, 6))
        sns.boxplot(data=df, x=column)
        plt.title(f'Boxplot of {column}')
        plt.xlabel(column)
        plt.show()

numerical_columns = df.select_dtypes(exclude=['bool']).columns

# Calculate mean and standard deviation before removing outliers
mean_before = df[numerical_columns].mean()
std_before = df[numerical_columns].std()

# Plot histograms and boxplots before removing outliers
plot_histograms(df)
plot_boxplots(df)
```

```python
# Remove the outliers from the dataset
lower_quantile = 0.01
upper_quantile = 0.99

# Print the shape of the original DataFrame
print("Shape before removing outliers:", df.shape)

# from the numerical columns just take (longitude, latitude, X Coordinate, Y Coordinate, IUCL, Description, Location Description, FBI Code)
numerical_columns = ['Longitude', 'Latitude', 'X Coordinate', 'Y Coordinate', 'IUCR', 'Description', 'Location Description', 'FBI Code']

# Remove outliers
for column in numerical_columns:
    if column != 'Primary Type':  # Exclude the target column if it's included
        lower_bound = df[column].quantile(lower_quantile)
        upper_bound = df[column].quantile(upper_quantile)
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove Null Values
df.dropna(inplace=True)
# Print the shape of the resulting DataFrame
print("Shape after removing outliers:", df.shape)
# df.info()
```
